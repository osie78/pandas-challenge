## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins=[0,9,14,19,24,29,34,39,100]
age_groups=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

age_df = df[["Age"]]
age_df["Age Groups"]=pd.cut(age_df["Age"], bins, labels=age_groups, include_lowest=True)

age_df = pd.DataFrame(age_df)
```

```python
age_df
```

```python
age_group_df = age_df.groupby("Age Groups").count()

age_group_df
```

```python
age_group_p_df=round(age_group_df/ all * 100,2)
```

```python
age_group_p_df=age_group_p_df.rename(columns={"Age":"Percentage of Players"})
```

```python
age_group_p_df = pd.DataFrame(age_group_p_df)
age_group_p_df
```

```python
bin_price_df = df[["Age","Price"]]

bin_price_df["Age Groups"]=pd.cut(bin_price_df["Age"], bins, labels=age_groups, include_lowest=True)
bin_price_df 
```

```python
bin_price_df
```

```python
bin_price_df[["Price", "Age Groups"]]
```

```python
bin_price_df=bin_price_df.groupby("Age Groups").sum()
bin_price_df=bin_price_df[["Price"]]
bin_price_df
```

```python
avg_bin_df = bin_price_df
avg_bin_df["Avg Purchase Price"]= bin_price_df["Price"] / age_group_df["Age"]
avg_bin_df = avg_bin_df["Avg Purchase Price"].round(2)
avg_bin_df =pd.DataFrame(avg_bin_df)
avg_bin_df
```

```python
bin_price_df=pd.DataFrame(bin_price_df)
bin_price_df
```

```python
resumen2=pd.concat([age_group_df,avg_bin_df,bin_price_df], axis=1)

resumen2
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
pop_p = df[["Item ID","Item Name", "Price"]]
pop_p = pop_p.groupby(["Item ID", "Item Name"]).count()
pop_p = pop_p.rename(columns={"Price":"Purchase Count"})
pop_p
```

```python
pop_price = df[["Item ID","Item Name", "Price"]]
pop_price = pop_price.groupby(["Item ID","Item Name"]).sum()
pop_price
```

```python
merge_pop_df1=pd.merge(pop_p,pop_price, on="Item Name")
merge_pop_df1=pd.DataFrame(merge_pop_df1)
merge_pop_df1
```

```python
pop_avg = df[["Item ID","Item Name", "Price"]]
pop_avg = pop_avg.groupby(["Item ID","Item Name"]).mean().round(2)
pop_avg
```

```python
merge_pop_df2=pd.merge(merge_pop_df1,pop_avg, on="Item Name")
merge_pop_df2=merge_pop_df2.rename(columns={"Price_x":"Total Purchase Value","Price_y":"Item Price"})
merge_pop_df2 = merge_pop_df2.sort_values("Purchase Count", axis = 0, ascending = False)
columns_titles = ["Purchase Count","Item Price","Total Purchase Value"]
merge_pop_df2=merge_pop_df2.reindex(columns=columns_titles)

merge_pop_df2_top5= merge_pop_df2.head()
merge_pop_df2_top5=pd.DataFrame(merge_pop_df2_top5)
merge_pop_df2_top5
```
